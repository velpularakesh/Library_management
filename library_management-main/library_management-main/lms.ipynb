```python
class LMS:
    def __init__(self,library_name,books_list):
        self.library_name = library_name
        self.books_list = books_list
        self.database = {}
        self.ID = 101
        db = open(self.books_list, 'r')
        books = db.readlines()
        db.close()
        for book in books:
            self.database.update({self.library_name[:2].upper()+(str(self.ID)):{'Book Title':book.replace('\n',''),'Lender Name':None,'Issue Date':None, 'Status':'Available'}})
            self.ID+=1

    def display_books(self):
        print('-'*28,'All Books','-'*28,'\n')
        print("Book ID".center(10),'Book Title'.center(30),'\t','Status'.center(10))
        print("-----------------------------------------------------------------------")
        for id,book in self.database.items():
            print(id.center(10),'\t',book['Book Title'].center(20),'\t\t',book['Status'])
    
    def display_available_books(self):
        print('---------------Available Books------------------')
        print("Book ID",'\t','Book Name')
        for id, book in self.database.items():
            if book['Status'] == 'Available':
                print(id,'\t\t',book['Book Title'])

    def issue_books(self):
        book_id = input("Enter book ID: ")
        if len(book_id)!=5:
            print("Enter valid book id")
            return self.issue_books()
        else:
            if book_id in self.database:
                if self.database[book_id]['Status'] == 'Available':
                    lender_name = input('Enter your name')
                    self.database[book_id]['Lender Name'] = lender_name
                    self.database[book_id]['Status'] = 'Unavailable'
                    print("Collect your book. Happy Learning!")
                else:
                    print("Books is unavailable")
                    return self.issue_books()
            else:
                print("Invalid book id")
                return self.issue_books()

    def return_books(self):
        book_id = input("Enter book id you want to return: ")
        if len(book_id)!=5:
            print("Enter valid book id")
            return self.return_books()
        else:
            if book_id in self.database:
                if self.database[book_id]['Status']=='Available':
                    print("The book is avaliable in the library. Please enter valid bood id")
                    return self.return_books()
                else:
                    self.database[book_id]['Status'] = 'Available'
                    print(f"Thank you { self.database[book_id]['Lender Name']}")
                    self.database[book_id]['Issue Date'] = ''
                    self.database[book_id]['Lender Name'] = ''
            else:
                print("Invalid book id")
                return self.return_books()

    def add_books(self):
        book_name = input("Enter the book name: ")
        if len(book_name)>15:
            print("Name should be in 15 characters")
            return self.add_books()
        elif len(book_name)==0:
            print("Enter valid name")
            return self.add_books()
        else:
            db = open(self.books_list,'a')
            db.writelines(f'\n{book_name}')
            db.close()
            self.database[self.library_name[:2].upper()+str(self.ID)]={'Book Title':book_name,'Lender Name':None,'Issue Date':None,'Status':'Available'}
            print("Book added successfully")

    def delete_book(self):
        pass
```

```python
library_1 = LMS("Central Library",'central_library.txt')
print(f"----------------Welcome to {library_1.library_name}---------------- ")
print("Choose an option below")
print("A. Display Books".center(10),"\nB. Display Available Books".center(30),"\nC. Issue Books".center(30),'\nD. Return Books'.center(30),'\nE. Add books'.center(30),'\nF. Quit')
option = False
while option!='F':
    option = input("Enter an option").upper()
    if option == 'A':
        print("\nDisplay Books")
        library_1.display_books()
    elif option== 'B':
        print("\nAvailable Books")
        library_1.display_available_books()
    elif option=='C':
        print("\nIssue Books")
        library_1.issue_books()
    elif option == 'D':
        print("\nReturn Books")
        library_1.return_books()
    elif option == 'E':
        print("\nAdd books")
        library_1.add_books()
    elif option == 'F':
        print("\nThank you! Visit again")
    else:
        print("\nSomething went wrong. Choose right option")
```
